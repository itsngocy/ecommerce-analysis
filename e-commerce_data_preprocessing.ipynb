## Cleaning

### Chuyn ct InvoiceDate v vi datatype datetime

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

### Missing Value

```python
df.isnull().sum()
```

### Drop missing value trong CustomerID v Description

```python
df = df.dropna(subset=['CustomerID', 'Description'])
df.isnull().sum()
```

### Xa Duplicate

```python
total_duplicates = df.duplicated().sum()

df.drop_duplicates(inplace=True)

print(f" xa {total_duplicates} dng duplicate.")
```

### Drop cc transaction Unitprice < 0

```python
df = df[df['UnitPrice'] > 0]
```

### To ct "Price"


```python
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
df.head(5)
```

### To ct InvoiceDateOnly

```python
# To ct ch gi li Ngy-Thng-Nm
df['InvoiceDateOnly'] = df['InvoiceDate'].dt.date
```

### X l cc stock code bt thng

```python
unique_stock_codes = df['StockCode'].unique()
anomalous_stock_codes = [code for code in unique_stock_codes if sum(c.isdigit() for c in str(code)) in (0, 1)]

print("Anomalous stock codes:")
for code in anomalous_stock_codes:
    print(code)

# xa cc stockcode l
df = df[~df['StockCode'].isin(anomalous_stock_codes)]
```

### X l description bt thng

```python
unique_Description = df['Description'].nunique()

top_10_Description = df['Description'].value_counts(normalize=True).head(10) * 100

print(f"S dng ca Description bt thng: {unique_Description}")

print(f"top 10 dng Description xut hin nhiu nht:")
print(top_10_Description)
```

X l cc dng trong description vit thng

```python
# tm cc dng vit thng
lowercase_descriptions = df['Description'].unique()
lowercase_descriptions = [desc for desc in lowercase_descriptions if any(char.islower() for char in desc)]

print("List cc gi tr trong ct Description c ch vit thng:")
for desc in lowercase_descriptions:
    print(desc)
```

```python
# xa cc dng cha Next Day Carriage v High Resolution Image
unique_value_descriptions = ["Next Day Carriage", "High Resolution Image"]
df = df[~df['Description'].isin(unique_value_descriptions)]
df.shape[0]
```

```python
# Chuyn sang kiu ch vit hoa
df['Description'] = df['Description'].str.upper()
```

## Cleaning

### Chuyển cột InvoiceDate về với datatype datetime

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

### Missing Value

```python
df.isnull().sum()
```

### Drop missing value trong CustomerID và Description

```python
df = df.dropna(subset=['CustomerID', 'Description'])
df.isnull().sum()
```

### Xóa Duplicate

```python
total_duplicates = df.duplicated().sum()

df.drop_duplicates(inplace=True)

print(f"Đã xóa {total_duplicates} dòng duplicate.")
```

### Drop các transaction Unitprice < 0

```python
df = df[df['UnitPrice'] > 0]
```

### Tạo cột "Price"


```python
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
df.head(5)
```

### Tạo cột InvoiceDateOnly

```python
# Tạo cột chỉ giữ lại Ngày-Tháng-Năm
df['InvoiceDateOnly'] = df['InvoiceDate'].dt.date
```

### Xử lý các stock code bất thường

```python
unique_stock_codes = df['StockCode'].unique()
anomalous_stock_codes = [code for code in unique_stock_codes if sum(c.isdigit() for c in str(code)) in (0, 1)]

print("Anomalous stock codes:")
for code in anomalous_stock_codes:
    print(code)

# xóa các stockcode lạ
df = df[~df['StockCode'].isin(anomalous_stock_codes)]
```

### Xử lý description bất thường

```python
unique_Description = df['Description'].nunique()

top_10_Description = df['Description'].value_counts(normalize=True).head(10) * 100

print(f"Số dòng của Description bất thường: {unique_Description}")

print(f"top 10 dòng Description xuất hiện nhiều nhất:")
print(top_10_Description)
```

Xử lý các dòng trong description viết thường

```python
# tìm các dòng viết thường
lowercase_descriptions = df['Description'].unique()
lowercase_descriptions = [desc for desc in lowercase_descriptions if any(char.islower() for char in desc)]

print("List các giá trị trong cột Description có chữ viết thường:")
for desc in lowercase_descriptions:
    print(desc)
```

```python
# xóa các dòng chứa Next Day Carriage và High Resolution Image
unique_value_descriptions = ["Next Day Carriage", "High Resolution Image"]
df = df[~df['Description'].isin(unique_value_descriptions)]
df.shape[0]
```

```python
# Chuyển sang kiểu chữ viết hoa
df['Description'] = df['Description'].str.upper()
```
